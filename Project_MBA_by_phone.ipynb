### Функции для рассчёта показателей

Support
Confidence
Expected confidence
Lift - данный показатель можно выразить через 2 другие функции, но для экономии вычислительного времени просто будем использовать полученные результаты и деление

```python
def support(a, b, df):
    counter = 0
    for i in range(0, len(df)):
        if df[a][i] > 0 and df[b][i] > 0:
            counter += 1
    return counter/len(df) 
```

```python
def confidence(a, b, df):
    counter = 0
    total = 0
    for i in range(0, len(df)):
        if df[a][i] > 0:
            total += 1 
            if df[b][i] > 0:
                counter += 1
    return counter/total 
```

```python
def expconf(a, b, df):
    counter = 0
    for i in range(0, len(df)):
        if df[b][i] > 0:
            counter += 1
    return counter/len(df) 
```

### Итоговая таблица

```python
for i in range(0, len(pairs)): 
    sup = support(pairs[i][0], pairs[i][1], MB) 
    conf = confidence(pairs[i][0], pairs[i][1], MB) 
    exconf = expconf(pairs[i][0], pairs[i][1], MB) 
    lift = (conf / exconf) 
    res['support'][i] = sup 
    res['confidence'][i] = conf 
    res['expected confidence'][i] = exconf 
    res['lift'][i] = lift 
```
